```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ="https://www.billboard.com/lists/best-latin-pop-songs-21st-century/me-gustas-tu-manu-chao-2001/"
response = requests.get(url)
soup = BeautifulSoup(response.text)
```

```python
response.status_code
```

```python
soup.select("article div div h2")
```

```python
artists = []

for artist in soup.select("h2.c-gallery-vertical-featured-image__title"):
    
    artists.append(artist.get_text(strip=True))
```

```python
artists
```

```python
artists = []
for artist in soup.select('article div div h2'):  # Placeholder selector
    artists.append(artist.get_text(strip=True).split(",")[0])
```

```python
artists
```

```python
songs = []

for song in soup.select('article div div h2'):  # Placeholder selector
    songs.append(song.get_text(strip=True).split(",")[1])  #.split()[0]
    
```
